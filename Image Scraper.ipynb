```python
import pandas as pd
import requests
import urllib2
import os, traceback
from base64 import b64decode
import csv
import base64
from httplib import BadStatusLine
```

```python
image_urls = pd.read_csv("Missing_Headphone_product_list - image.csv")
```

```python
image_urls
```

```python
def create_dir(dirname):
    try:
        if not os.path.exists(dirname):
            os.makedirs(dirname)
    except Exception as Ex:
        print (Ex)
        pass
```

```python
output = []
k = 1
opid = None

image_urls = image_urls.sort_values(by=['product_id'])[['product_id', 'product_name', 'image_loc']]
for row in image_urls.to_dict('records'):
    pid = row['product_id']
    pname = row['product_name']
    im_url = row['image_loc'].replace('webp','jpeg')
    parent_dir = './'
    
    if pid != opid:
        k = 1
        opid = pid
    try:
        create_dir(parent_dir+str(pid))
    except Exception as Ex:
        print (Ex)
        pass
#     image_name_f = str(pid)+'/'+str(pname).strip().replace(" ","-").replace("/","---").replace('&','-and-').replace('+','-plus-').replace(',','-')+'-{}.jpeg'
    image_name_f = str(pid)+'/'+str(pid)+'-'+'p'+'-{}.jpeg'
    image_name = image_name_f.format(k)
    fp = open(parent_dir+image_name,'wb+')
    if im_url.startswith('data') and 'base64' in im_url:
        fp.write(im_url.split(',')[-1].decode('base64'))
    elif 'http' in im_url or 'https' in im_url:
        try:
            req = urllib2.Request(im_url)
            req.add_header('User-Agent','Mozilla/4.0 (compatible; MSIE 8.0; Windows NT 6.0; Trident/4.0; Acoo Browser; GTB5; Mozilla/4.0 (compatible; MSIE 6.0; Windows NT 5.1; SV1) ; InfoPath.1; .NET CLR 3.5.30729; .NET CLR 3.0.30618')
            fp.write(urllib2.urlopen(req).read())
        except Exception as Ex:
            print (traceback.format_exc())
            continue
        
#     product_image.write('~'.join([str(pid),image_name, status,'\n']))
    row['image_loc'] = image_name
    
    if k == 1:
        row['image_type'] = 'P'
    else:
        row['image_type'] = 'S'
    
    output.append(row)
    k += 1
# product_image.close()
```

```python
image_output = pd.DataFrame(output)
```

```python
image_output.to_csv("headphone_images_new_today.csv",index=False)
```
